```python
import librosa
import librosa.display
from glob import glob

# Load audio file
extract_path = "../data/external/"
dataset_path = glob(extract_path + "*/")[0]
sample_rate = 48000
audio0 = librosa.load(dataset_path + "samePhrase/1/1-1.flac", sr = sample_rate)[0]
audio0
```

```python
# Inspect audio waveform
librosa.display.waveshow(audio0)
```

```python
import matplotlib.pyplot as plt

# Generate mel spectrogram
spec0 = librosa.feature.melspectrogram(y = audio0, sr = sample_rate)
librosa.display.specshow(spec0, x_axis = "s", y_axis = "mel", sr = sample_rate)
plt.colorbar()
```

```python
import numpy as np

# Convert to dB
spec0 = librosa.power_to_db(spec0, ref = np.max)
librosa.display.specshow(spec0, x_axis = "s", y_axis = "mel", sr = sample_rate)
plt.colorbar()
```

```python
# Normalise
spec0 = (spec0 - spec0.mean()) / spec0.std()
spec0 = (spec0 - spec0.min()) / (spec0.max() - spec0.min())
spec0.shape
```
